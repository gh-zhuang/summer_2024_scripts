```python
## basic
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

## statistics
from scipy.stats import zscore

## system
import os
import pickle
import warnings
warnings.filterwarnings("ignore")

##
import umap.umap_ as umap
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import KernelPCA
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

## Define Helper Functions

```python
def scale(df,
          method="standard"):
    if method == "standard":
        scaler = StandardScaler()
        df_scaled = pd.DataFrame(scaler.fit_transform(df),
                                 index = df.index,
                                 columns = df.columns)
        return df_scaled

    if method == "double_z":
        df_scaled = df.apply(zscore,axis=1).\
            apply(zscore, axis=0)

    else:
        df_scaled = df

    return df_scaled

def umap_wrapper(df,
                 df_meta,
                 pca=True,
                 n_pca=10,
                 remove_PC1=True,
                 n_neighbors=15,
                 n_umap=3,
                 learning_rate=1,
                 min_dist=0.1,
                 spread = 1,
                 metric="euclidean",
                 random_state=42):

    assert min_dist <= spread

    print(f"Input data dimension: {df.shape}")
    print(f"Input meta-data dimension: {df_meta.shape}")
    
    ## perform linear PCA
    pca = PCA(n_components=n_pca)
    df_pca = pd.DataFrame(pca.fit_transform(df),
                          columns = ["PC" + str(i) for i in range(1, n_pca+1)],
                          index = df.index)

    ## remove "PC1"
    if remove_PC1:
        df_pca = df_pca.drop("PC1",axis=1)
    
    print(f"Dimension after PCA: {df_pca.shape}")

    ## UMAP
    umap_embedding = umap.UMAP(n_neighbors=n_neighbors,
                               n_components=n_umap, 
                               learning_rate=learning_rate,
                               min_dist=min_dist,
                               spread = spread,
                               metric=metric,
                               random_state=random_state)
    
    res = pd.DataFrame(umap_embedding.fit_transform(df_pca),
                       columns = ["UMAP" + str(i) for i in range(1, n_umap+1)],
                       index = df_pca.index)

    print(f"Dimension after UMAP: {res.shape}")
    
    ## merge meta-data
    res = pd.merge(df_meta, res, left_index=True, right_index=True)

    print(f"Dimension after merging: {res.shape}")

    return(res)
```

## Load Data

```python
with open('/screening/notebooks/zhuang/summer_2024/data/region_ids.pkl', 'rb') as f:
        region_ids = pickle.load(f)

with open('/screening/notebooks/zhuang/summer_2024/data/df_ldt.pkl', 'rb') as f:
        df_ldt = pickle.load(f)

## Query region meta-data to obtain "positive_control" regions
v6_region = "/screening/data/methylation/sandbox_pipeline/assets/mafband_v2023_12_23/msre_regions.tsv"
pos_control = pd.read_csv(v6_region, sep="\t").\
    query('merged_region_type == "HYPER_CONTROL"').\
    loc[:,"region_id"].\
    reset_index(drop=True)
pos_control_in_s3 = [i for i in pos_control if i in region_ids]

print(f"There are {len(pos_control_in_s3)} positive control regions in S3")

## subset data
df_lung = df_ldt.loc[:,["unique_sample_id","sample_group","histological_subtype", 'g360_max_maf_pct', 'tvf_max_maf_pct']+region_ids].\
    query('sample_group == "lung"').\
    drop("sample_group",axis=1).\
    set_index("unique_sample_id").\
    drop(pos_control_in_s3,axis=1).\
    assign(histological_subtype = lambda df: np.where(df['histological_subtype'].isin(['adenocarcinoma', 'squamous_cell_carcinoma']), df['histological_subtype'], "unknown"),
          log_g360_maf = lambda df: np.log(df['g360_max_maf_pct']),
          log_tvf_maf =  lambda df: np.log(df['tvf_max_maf_pct']))
        
#query('histological_subtype == "adenocarcinoma"')
#.\
    #query(' g360_max_maf_pct <= 1 ')

print(df_lung.shape)
```

## Kernel PCA

```python
kernels = ['linear','rbf','cosine','poly']
pdata = ['histological_subtype', 'g360_max_maf_pct', 'tvf_max_maf_pct', 'log_g360_maf', 'log_tvf_maf']
bins = 5

## meta-data
df_meta = df_lung.loc[:,pdata].\
    assign(log_g360_maf_binned = lambda df: pd.cut(df['log_g360_maf'],bins,labels=["bin " + str(i) for i in range(1,bins+1)]),
           log_tvf_maf_binned = lambda df: pd.cut(df['log_tvf_maf'],bins,labels=["bin " + str(i) for i in range(1,bins+1)]))

## region score and scaling
df = df_lung.drop(pdata,axis=1)
df_standard = scale(df,
                    method = "standard")
df_double_z = scale(df,
                    method = "double_z")

print(df.shape)
```

### Standard Scaling

```python
data = df_standard
hue = 'log_g360_maf_binned'
n = 10
x = "PC1"
y = "PC2"

fig, axes = plt.subplots(2,2,figsize=(12,10))

axes = axes.flatten()

with plt.style.context('_classic_test_patch'):
    for idx, kernel in enumerate(kernels):
        
        res = KernelPCA(n_components = n, kernel = kernel).fit_transform(data)
        
        PCs = ["PC" + str(i) for i in range(1,n+1)]
        
        res = pd.DataFrame(res, 
                           index = data.index,
                           columns=PCs)
        res = pd.merge(df_meta, res, left_index=True, right_index=True)
        print(res.shape)
        
        sns.scatterplot(x = x,
                        y = y,
                        hue = hue,
                        data = res,
                        ax = axes[idx])
        
        axes[idx].set_title("lung+"+kernel+"+standard", fontsize=12)
```

### Double-z

```python
data = df_double_z
hue = 'log_g360_maf_binned'
n = 10
x = "PC1"
y = "PC2"

fig, axes = plt.subplots(2,2,figsize=(12,10))

axes = axes.flatten()

with plt.style.context('_classic_test_patch'):
    for idx, kernel in enumerate(kernels):
        
        res = KernelPCA(n_components = n, kernel = kernel).fit_transform(data)
        
        PCs = ["PC" + str(i) for i in range(1,n+1)]
        
        res = pd.DataFrame(res, 
                           index = data.index,
                           columns=PCs)
        res = pd.merge(df_meta, res, left_index=True, right_index=True)
        print(res.shape)
        
        sns.scatterplot(x = x,
                        y = y,
                        hue = hue,
                        data = res,
                        ax = axes[idx])
        
        axes[idx].set_title("lung+"+kernel+"+double_z", fontsize=12)
```

## UMAP

- Use standard scaling or double-z scores
- Perform linear PCA and use the top 10-20 PCs
- Perform UMAP in the PC-space

```python
res = umap_wrapper(df_standard,
                   df_meta,
                   n_pca=20,
                   remove_PC1=True,
                   n_neighbors=10,
                   n_umap=3,
                   learning_rate=0.1,
                   min_dist=0.5,
                   spread=1,
                   metric="euclidean",
                   random_state=42)
```

```python
hue = "log_g360_maf"

fig, axes = plt.subplots(figsize=(8,6))
with plt.style.context('_classic_test_patch'):
        
    sns.scatterplot(x = "UMAP1",
                    y = "UMAP2",
                    hue = hue,
                    data = res,
                    ax = axes)
        
    axes.set_title("lung+standard", fontsize=12)
    axes.set_xlabel("UMAP 1", fontsize=10)
    axes.set_ylabel("UMAP 2", fontsize=10)
```
